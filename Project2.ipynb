```python
#9 en küçükten en büyüğe doğru görüntülenme değerlerini sıralayınız ve ilk 10 kaydı alınız.

#1.yol
print(c.sort_values("views",ascending=False)[["title","views"]].head(10),"\n\n")  #sort_values fonksiyonunda ascending belirtilmezse True olarak varsayacaksın. Yani değerler Yukarıdan aşağıya gidecek şekilde küçükten büyüğe doğru ilerler.
#2.yol
c.sort_values("views").tail(10)
```

```python
#10 kategoriye göre beğeni ortalamalarını yukarıdan aşağıya küçükten büyüğe doğru sıralayınız.
c.groupby("category_id")[["likes"]].mean().sort_values("likes")
```

```python
#11 kategoriye göre yorum sayılarını yukarıdan aşağıya büyükten küçüğe doğru sıralayınız.
import numpy as np
d = c.groupby("category_id")[["comment_count"]].sum().sort_values("comment_count",ascending=False)
d
```

```python
#13 her videonun title uzunluğu bilgisini yeni bir kolonda gösteriniz.
c["title_len"] = c["title"].apply(lambda x: len(x))   # buradaki olay apply metoduyla her title bilgisini apply metodu aracılığıyla len fonksiyonuna göndermek
c
```

```python
#14 her video için kullanılan tag sayısını yeni kolonda gösteriniz.
c["tag_count"] = c["tags"].apply(lambda x: len(x.split("|")))
c
```

```python
sns.scatterplot(x = "likes",y = "dislikes",hue="category_id",data=c);
```
