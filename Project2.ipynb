```python
#8 en çok görüntülenen video hangisidir ?

df[df["views"]==df["views"].max()][["title","views"]]
```

```python
#9 en düşük görüntülenen video hangisidir ?

df[df["views"]==df["views"].min()][["title","views"]]
```

```python
#10 en küçükten en büyüğe doğru görüntülenme değerlerini sıralayınız ve ilk 30 kaydı alınız.

df.sort_values("views",ascending=False)[["title","views"]].head(30)  #sort_values fonksiyonunda ascending belirtilmezse True olarak varsayacaksın. Yani değerler Yukarıdan aşağıya gidecek şekilde küçükten büyüğe doğru ilerler.
```

```python
#11 kategoriye göre like ve dislike oranlarının ortalamasını gösteriniz.

df.groupby("category_id")[["likes","dislikes"]].aggregate(np.mean)
```

```python
#12 kategoriye göre yorum sayılarını yukarıdan aşağıya büyükten küçüğe doğru sıralayınız.

df.groupby("category_id")[["comment_count"]].sum().sort_values("comment_count",ascending=False)
```

```python
#13 her kategoride kaç video vardır ?

df.groupby("category_id").count()[["title"]]
```

```python
df["category_id"].value_counts()
```

```python
df.groupby("category_id").count()
```

```python
#14 her videonun title uzunluğu bilgisini yeni bir kolonda gösteriniz.
df["title_len"] = df["title"].apply(lambda x: len(x))   # buradaki olay apply metoduyla her title bilgisini apply metodu aracılığıyla len fonksiyonuna göndermek
df
```

```python
#15 her video için kullanılan tag sayısını yeni kolonda gösteriniz.
df["tag_count"] = df["tags"].apply(lambda x: len(x.split("|")))
df
```

```python
#16
sns.scatterplot(x = "likes",y = "dislikes",hue="category_id",data=df);
```
