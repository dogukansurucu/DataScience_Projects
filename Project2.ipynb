```python
#11 kategoriye g√∂re like ve dislike oranlarƒ±nƒ±n ortalamasƒ±nƒ± g√∂steriniz.

df.groupby("category_id")[["likes","dislikes"]].aggregate(np.mean)
```

```python
#12 kategoriye g√∂re yorum sayƒ±larƒ±nƒ± yukarƒ±dan a≈üaƒüƒ±ya b√ºy√ºkten k√º√ß√ºƒüe doƒüru sƒ±ralayƒ±nƒ±z.

df.groupby("category_id")[["comment_count"]].sum().sort_values("comment_count",ascending=False)
```

```python
#13 her kategoride ka√ß video vardƒ±r ?

df.groupby("category_id").count()[["title"]]
```

```python
#14 her videonun title uzunluƒüu bilgisini yeni bir kolonda g√∂steriniz.
df["title_len"] = df["title"].apply(lambda x: len(x))   # buradaki olay apply metoduyla her title bilgisini apply metodu aracƒ±lƒ±ƒüƒ±yla len fonksiyonuna g√∂ndermek
df
```

```python
#15 her video i√ßin kullanƒ±lan tag sayƒ±sƒ±nƒ± yeni kolonda g√∂steriniz.
df["tag_count"] = df["tags"].apply(lambda x: len(x.split("|")))
df
```
